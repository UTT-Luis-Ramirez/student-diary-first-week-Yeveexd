#PROBABILIDAD Y ESTADISTICA

**axiomas:** afirmaciones o negaciones que son correctas hasta que se demuestre lo contrario
**falacias:** afirmaciones que son incorrectas hasta que se demuestre lo contrario

la **estadistica** es una rama de las matematicas que se encarga de la recoleccion, organizacion, analisis e interpretacion de datos

##***estadistica descriptiva:*** la estadistica descriptiva es una rama de la estadistica que se encarga en resumir y describir las caracteristicas principales de una conjunto de datos. su objetivoes organizar, presentar y analizar la informacion de manera mas facil de comprender e interpretar.
 busca describir los datos utilizando medidas como medidas centrales, dispercion y graficas

 para comprender mejor esta definicion podemos desglosarla en sus funciones:

 1.-***recoleccion y organizacion de datos:*** la estadistica descriptiva comienza con la recoleccion de datos relevantes para el estudio. luego, estos daros se organizan en tablas, graficos y otras formasque faciliten su analista

 2.- ***calculo de medidas descriptivas:*** se utilizan para resumir los datos, como medidas de tendencia central (media, mediana, moda), medidas de dispersion (desviacion estandar, rango, varianza) y medidas de forma (asimetria, curtosis)

 3.- ***presentacion de datos:*** los datos y las medidas descriptivas se presentan de forma visualmente atractiva, y facil de entender, utilizando tablas, graficos historigramas, diagramas de caja y otras herramientas.

 4.- ***identificacion de patrones y relaciones:*** la estadistica descriptiva permiten las relaciones entre las variables de un conjunto de datos, esto puede incluir la deteccion de tendencias, la comparacion de grupos y exploracion de correlacuones.


##***estadisticas inferencial:***
se basa en habilidades o metodos de probabilidad en diferentes tipos de muestras, probabilidad, muestre e hipotesis.
la estadistica se aplica en investigaciones, economia, medicina y ciencias sociales
para comprender mejor esta definicion, podemo desglosarla en sus funciones principales:

1;- ***estrimacioon de parametros:*** se utilizan metodos estadisticos para estimar valores de parametros desconocidos de la poblacion, como la media, la desviacion estandar o la proporcion.

2.- ***pruebas de hipotesis:*** se formuklan hipotesis sobre la poblacion y se utilizan datos muestrales para evaluar evidencia a favor o en contra de las hipotesis

3.- ***intervalos de confianza:*** se construyen intervalos de valores que tienen una alta probabilidad de contener el verdadero valor del parametro poblacional

4.- ***modelado  estadistico:*** se utilizanmodelos matematicos para representar la relacion entre variables en la poblacion. estos modelos prermitren realizar predicciones sobre el comportamiento futuro de la poblacion y corprender las interacciones entre las variables.


 ## Análisis del Comportamiento del Usuario

###*Observación Inicial:*
- Un sitio e-commerce registra las páginas vistas y los clics.
- El 60% de los usuarios abandonan en la página del carrito de compras.

###*Identificación del Problema:*
- La alta tasa de abandono en la página del carrito de compras indica que los usuarios están experimentando dificultades o frustraciones en esta etapa del proceso de compra.

###*Hipótesis:*
- La página del carrito de compras puede estar siendo confusa, complicada o no intuitiva, lo que lleva a los usuarios a abandonar antes de completar la compra.

###*Solución Implementada:*
- Se rediseñó la página de checkout para hacerla más intuitiva y fácil de usar.

###*Resultados:*
- Después de implementar el nuevo diseño, la tasa de conversión aumentó en un 25%.
